```python
class Product:
  def __init__(self, product_id, name, price, stock):
    self.product_id = product_id
    self.name = name
    self.price = price
    self.stock = stock

  def __str__(self):
    return f"Product ID: {self.product_id}\nName: {self.name}\nPrice: {self.price}\nStock: {self.stock}"
```

```python
class User:
  def __init__(self, user_id, username, email):
    self.user_id = user_id
    self.username = username
    self.email = email

  def __str__(self):
    return f"User ID: {self.user_id}\nUsername: {self.username}\nEmail: {self.email}"
```

```python
class Order:
  def __init__(self, order_id, user_obj, items, total_price, status):
    self.order_id = order_id
    self.user_obj = user_obj
    self.items = items
    self.total_price = total_price
    self.status = status

  def __str__(self) -> str:
    order_str = f"Order ID: {self.order_id}\nUser:\n{self.user_obj}\nItems:\n"
```

```python
class Item:
  def __init__(self, product_obj, quantity):
    self.product_obj = product_obj
    self.quantity = quantity

  def __str__(self):
    return f"Product: {self.product_obj.name}, Quantity: {self.quantity}"
```
