```python
def main():
    Number_of_Orders = [100, 20, 100, 4500, 500]
    k = 3
    ans = 0



    curr_sum = sum(Number_of_Orders[:k])
    ans = curr_sum
    st = 0


    for j in range(k, len(Number_of_Orders)):
        curr_sum += Number_of_Orders[j]
        curr_sum -= Number_of_Orders[j - k]
        # ans = max(ans, curr_sum)

        if curr_sum > ans:
          ans = curr_sum
          st = j - k + 1


    print(" Max order Window Starts:", st)
    return ans


if __name__ == "__main__":
   print(main())
```

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        l=0
        r=len(s)-1

        while(l<r):

            #remove non alphanumnfrom left
            while(l<r and not s[l].isalnum()):
                l += 1

            #same from right
            while(l<r and not s[r].isalnum()):
                r -= 1


            if s[l].lower() != s[r].lower():
                return False
            l += 1
            r -= 1

        return True
```

```python
s = "A man, a plan, a canal: Panama"
obj = Solution()
print(obj.isPalindrome(s))
```
