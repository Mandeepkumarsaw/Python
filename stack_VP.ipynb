```python
class ValidParenthesis:
    def is_valid(self, s: str) -> bool:
        stack = []
        for c in s:
            if c in ['{', '(', '[']:
                stack.append(c)
            elif c in ['}', ')', ']']:   #only act on closing brackets
                if not stack:
                    return False
                if (c == ')' and stack[-1] != '(') or \
                   (c == '}' and stack[-1] != '{') or \
                   (c == ']' and stack[-1] != '['):
                    return False
                stack.pop()
            # else: ignore all other characters
        return not stack


# Example usage
vp = ValidParenthesis()
api_payload_1 = '{"order": {"id": 1001, "items": [1, 2, 3]}}'
api_payload_2 = '{"order": {"id": 1001, "items": [1, 2, 3]}'

print(vp.is_valid(api_payload_1))  # True
print(vp.is_valid(api_payload_2))  # False
```
