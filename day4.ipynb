```python
import requests

def func():
    response = requests.get("https://fake-store-api.mock.beeceptor.com/api/products")
    data = response.json()

    category_count = {}

    for product in data:
        category = product.get("category")

        if category in category_count:
            category_count[category] += 1
        else:
            category_count[category] = 1

    print(category_count)

func()
```

```python
#TASK-3 For each product, find out the number of orders made on that product
```

```python
import requests

def product_order_count():
    orders = requests.get("https://fake-store-api.mock.beeceptor.com/api/orders").json()

    product_count = {}

    for order in orders:
        for item in order.get("items", []):
            pid = item["product_id"]
            qty = item.get("quantity", 0)  # get quantity

            if pid in product_count:
                product_count[pid] += qty
            else:
                product_count[pid] = qty

    sorted_dict = dict(sorted(product_count.items()))
    print(sorted_dict)
    # print(product_count)

product_order_count()
```

```python
import requests

def func():
    products = requests.get("https://fake-store-api.mock.beeceptor.com/api/products").json()
    orders = requests.get("https://fake-store-api.mock.beeceptor.com/api/orders").json()

    # Step 1: count orders per product
    order_count = {}
    for order in orders:
        for item in order.get("items", []):
            pid = item["product_id"]
            order_count[pid] = order_count.get(pid, 0) + 1

    # Step 2: calculate avg rating per category
    category_rating = {}
    category_count = {}

    for product in products:
        category = product.get("category")
        rating = product.get("rating", 0)

        category_rating[category] = category_rating.get(category, 0) + rating
        category_count[category] = category_count.get(category, 0) + 1

    # avg = total / count
    category_avg = {}
    for cat in category_rating:
        category_avg[cat] = category_rating[cat] / category_count[cat]

    # Step 3: build final result
    result = []

    for product in products:
        pid = product.get("product_id")
        category = product.get("category")

        result.append({
            "product_id": pid,
            "product_name": product.get("name"),
            "rating": round(category_avg[category], 2),  # avg rating of category
            "No_of_orders": order_count.get(pid, 0)
        })

    print(result)

func()
```
