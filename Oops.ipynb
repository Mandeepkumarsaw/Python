```python
import requests

class Product:
  def __init__(self,product_id,name,price):
    self.product_id = product_id
    self.name = name
    self.price = price

  def printD(self):
    print("Product_id: ",self.product_id)
    print("Name: ",self.name)
    print("Price: ",self.price,"\n")

response = requests.get("https://fake-store-api.mock.beeceptor.com/api/products")
# print(response)
products = response.json()

# product_list = []

for items in products:
  obj = Product(
       items["product_id"],
       items["name"],
       items["price"]
  )

  obj.printD()

  # product_list.append(obj)

# for i in product_list:
#   print(i.product_id,i.name,i.price)



```
