```python
import requests

data = requests.get("https://fakestoreapi.com/users/").json()


def flatten(data, parent_key="", result=None):

    if result is None:
        result = {}

    for key, value in data.items():

        new_key = f"{parent_key}.{key}" if parent_key else key

        if isinstance(value, dict):
            flatten(value, new_key, result)

        else:
            result[new_key] = value

    return result


# API returns list of users
for i in data:
    print(flatten(i))
    print()
```

```python
product_record = {
    "product_id": 101,
    "name": "Laptop",
    "pricing": {
        "base_price": 80000,
        "discount": {
            "percentage": 10,
            "active": True
        }
    },
    "inventory": {
        "warehouse": {
            "city": "Bangalore",
            "stock": 25
        }
    }
}


def flatten_dict(data, parent_key="", result=None):

    # create result dict only first time
    if result is None:
        result = {}

    for key, value in data.items():

        # create new key
        new_key = f"{parent_key}.{key}" if parent_key else key

        # if value is dict → recurse again
        if isinstance(value, dict):
            flatten_dict(value, new_key, result)

        # base condition
        else:
            result[new_key] = value

    return result


output = flatten_dict(product_record)

print(output)
```

```python
import requests


def flatten_dict(url):

    data = requests.get(url).json()


    for item in data:

        result = {}

        def flatten(d, parent_key=""):

            for key, value in d.items():

                new_key = f"{parent_key}.{key}" if parent_key else key

                # recursion
                if isinstance(value, dict):
                    flatten(value, new_key)

                else:
                    result[new_key] = value

        # send one dictionary at a time
        flatten(item)

        print(result)
        print()


# function call
flatten_dict("https://fakestoreapi.com/users/")
```
