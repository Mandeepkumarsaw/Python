```python
import requests
import time

target_emails = []

while len(target_emails) < 10:
    response = requests.get("https://randomuser.me/api/")
    data = response.json()

    user = data['results'][0]

    name = user['name']['first'] + " " + user['name']['last']
    age = user['dob']['age']
    gender = user['gender']
    email = user['email']

    if gender == "female" and age < 25:
        target_emails.append(email)
        print("Collected:", email)
    else:
        print("No match")

    time.sleep(1)

print("\nFinal Emails:")
for e in target_emails:
    print(e)
```

```python
import requests

target_emails = set()

while len(target_emails) < 10:
    # Get multiple users in one request (important)
    response = requests.get("https://randomuser.me/api/?results=10")
    data = response.json()

    for user in data['results']:
        # Extract fields
        name = user['name']['first'] + " " + user['name']['last']
        age = user['dob']['age']
        gender = user['gender']
        email = user['email']

        # Apply filter
        if gender == "female" and age < 25:
            if email not in target_emails:
                target_emails.add(email)
                print(f"Collected: {name}, {age}, {email}")

        # Stop when 10 collected
        if len(target_emails) == 10:
            break

    print("Total collected so far:", len(target_emails))
    print("-" * 40)

# Final Output
print("\nFinal Emails:")
for e in target_emails:
    print(e)
```
