```python
# DO NOT MAKE ANY CHANGES HERE !!!!!
import random
import time
def generate_trade():
    symbols = ["AAPL", "TSLA", "GOOG"]
    sides = ["BUY", "SELL"]

    trade = {
        "trade_id": random.randint(1000, 9999),
        "symbol": random.choice(symbols),
        "side": random.choice(sides),
        "quantity": str(random.randint(1, 100)),
        "price": str(round(random.uniform(100, 500), 2))
    }

    return trade
```

```python
while True:
  trade = generate_trade()
  print(trade)
  time.sleep(2)
```

USE the generate_trade method to receive trades and process the trades for next 1 minute to find out the P&L
