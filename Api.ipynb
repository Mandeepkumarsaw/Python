```python
import requests

def check_product_delivered(product_id):
    url = "https://fake-store-api.mock.beeceptor.com/api/orders"
    data = requests.get(url).json()

    found = False

    for order in data:
        for item in order['items']:
            if item['product_id'] == product_id:
                found = True
                if order['status'] == "Delivered":
                    print(f"Product {product_id} → Delivered  (Order {order['order_id']})")
                else:
                    print(f"Product {product_id} → Not Delivered  (Order {order['order_id']})")

    if not found:
        print(f"Product {product_id} not found in any order")
```

```python
import requests

def giveProductPrice(product_id):
    response = requests.get("https://fake-store-api.mock.beeceptor.com/api/products/")
    datas = response.json()

    for product in datas:
        if product['product_id'] == product_id:
            return product['price']
    return 0


def calculateCartValue():
    response = requests.get("https://fake-store-api.mock.beeceptor.com/api/carts")
    carts = response.json()

    total = 0

    for cart in carts:
        for item in cart['items']:
            product_id = item['product_id']
            quantity = item['quantity']

            price = giveProductPrice(product_id)
            total += price * quantity

    return total


print("Total Cart Value:", calculateCartValue())
```
