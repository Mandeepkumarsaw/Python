```python
class Account:
  def __init__(self,name,age,accnu,bal):
    self.name = name
    self.age = age
    self.accnu = accnu
    self.bal = bal

  def get_info(self):
      print(f"Name: {self.name} | Age: {self.age}, | Bal: {self.bal}")



class SavingAcc(Account):
  def __init__(self,name,age,accnu,bal,acctype,Interest,FDcredential):
    super().__init__(name, age, accnu, bal)
    self.acctype = acctype
    self.Interest = Interest
    self.__FDcredential = FDcredential   #Make it private
    # print(self.__FDcredential)

  def get_fd(self):
        return self.__FDcredential

  def sav_info(self):
      print(f"Account Type: {self.acctype}, Rate of Interest: {self.Interest}")



class CurrAcc(SavingAcc):
  def __init__(self,name,age,accnu,bal,acctype,Interest,limit,FDcredential):
    super().__init__(name,age,accnu,bal,acctype,Interest,FDcredential)
    self.limit = limit

  def curr_info(self):
      print(f"OverDraft Limit: {self.limit}, FDcredential: {self.get_fd()}")

```

```python
obj1 = CurrAcc("Ram",21,111222333,50000,"Saving Acc",6.5,500000,"FD123456")
print(obj1.__dict__)


# print(obj1.__FDcredential)              # Attribute ERROR ---> benefits of making private
print(obj1._SavingAcc__FDcredential)      # However can be access using class-->var(eventhough it's private)
```

```python
obj1.get_info()
obj1.sav_info()
obj1.curr_info()

# print(obj1.get_fd())              # can be access using Function as it declared inside the scope of that private variable
```
