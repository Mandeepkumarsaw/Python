```python
import requests


data = requests.get(
    "https://fake-store-api.mock.beeceptor.com/api/products").json()

#Sort
sorted_data = sorted(data, key=lambda x: x['rating'], reverse=True)


result = []

for item in sorted_data:
    result.append({
        "product_id": item["product_id"],
        "best_review_user_id": item["rating"]
    })

print(result)
```

```python
import requests


data = requests.get("https://fakestoreapi.com/products/").json()

sorted_rating = sorted(data, key=lambda x:x['rating']['rate'],reverse = True)
# print(sorted_rating)

result = []
for item in sorted_rating:
  result.append({"Title" : item['title'],
                "Top rating" : item['rating']['rate']})

  print(result)
```
