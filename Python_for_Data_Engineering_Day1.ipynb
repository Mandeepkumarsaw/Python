## Read **raw** data (`list` of `dict`)

**Keywords:** **raw** В· **dirty** В· **`list`** В· **`dict`** В· **record** / row

Below: fake **extract** вҖ” one bad **amount** on purpose (**invalid** for our rule later).

````python
# Raw extract вҶ’ Python structures (keywords: list, dict, keys, values)
orders = [
    {"id": 1, "amount": 250},
    {"id": 2, "amount": -100},
    {"id": 3, "amount": 450},
]

print("raw_orders:", orders)
print("len(raw):", len(orders))
````

## `print` and **control flow**

**Keywords:** **`print()`** В· **statement** В· runs **top to bottom** (control flow)

````python
# Sequential execution вҖ” keyword: print
print("step_1_start")
total_before_cleaning = 250 + (-100) + 450
print("naive_sum_includes_bad_row:", total_before_cleaning)
print("step_3_end")
````

## **Clean** with **`for`** + **`if`** + **`.append()`**

**Keywords:** **filter** В· **validate** В· **`for`** В· **`if`** В· **`>`** В· **`.append()`** В· **`clean_orders`**

**Rule:** keep only `order["amount"] > 0` (**drop** negatives).

````python
# Transform: raw list вҶ’ clean list (same pattern as tiny ETL "T" step)
clean_orders = []
for order in orders:
    if order["amount"] > 0:
        clean_orders.append(order)

print("clean_orders:", clean_orders)
print("len(raw):", len(orders), "len(clean):", len(clean_orders))
````

## What happened (**keywords**)

- **Input:** **`orders`** = **raw** / **dirty**  
- **Output:** **`clean_orders`** = **validated** subset  
- **Pipeline idea:** **raw вҶ’ clean** (first **trust** step before **aggregate** / **load**)
- **Pattern:** **ETL** вҖ” here: **extract** (list above) вҶ’ **transform** (loop) вҶ’ **load** (not shown; would be file/API/warehouse)
