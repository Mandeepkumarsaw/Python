```python
def binarySearch(items,target):
  left = 0
  right = len(items)-1

  while left <= right:
    mid = left + (right - left) // 2

    if items[mid] == target:
      return mid
    elif items[mid] < target:
      left = mid + 1
    else:
      right = mid - 1

  return -1
```
