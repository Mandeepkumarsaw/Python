```python
#TASK 1--> TOTAL CART VALUE
```

```python
import requests

def giveProductPrice(product_id):
    response = requests.get("https://fake-store-api.mock.beeceptor.com/api/products/")
    datas = response.json()

    for product in datas:
        if product['product_id'] == product_id:
            return product['price']
    return 0


def calculateCartValue():
    response = requests.get("https://fake-store-api.mock.beeceptor.com/api/carts")
    carts = response.json()

    total = 0

    for cart in carts:
        for item in cart['items']:
            product_id = item['product_id']
            quantity = item['quantity']

            price = giveProductPrice(product_id)
            total += price * quantity

    return total


print("Total Cart Value:", calculateCartValue())
```

```python
#TASK-2 Validate with Price given by order API & checked the Status
```

```python
import requests

def validate_cart_with_orders():
    cart_total = calculateCartValue()

    orders = requests.get("https://fake-store-api.mock.beeceptor.com/api/orders").json()

    for order in orders:
        api_total = order['total_price']
        status = order['status']
        order_id = order['order_id']

        print(f"\nOrder ID: {order_id}")

        #compare totals
        if cart_total == api_total:
            print("Cart value matches order total")
        else:
            print("Mismatch")
            print("Cart:", cart_total, "Order:", api_total)

        #status
        if status.lower() == "delivered":
            print("Status: Delivered")
        else:
            print("Not Delivered:", status)


validate_cart_with_orders()
```
