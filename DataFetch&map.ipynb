```python
import requests

class Product:
    def __init__(self, product_id, title, price, category):
        self.product_id = product_id
        self.title = title
        self.price = price
        self.category = category

    def __str__(self):
        return f"{self.title} (₹{self.price})"


class User:
    def __init__(self, id, username, email):
        self.id = id
        self.username = username
        self.email = email

    def __str__(self):
        return f"{self.username} ({self.email})"


class Item:
    def __init__(self, product_obj, quantity):
        self.product_obj = product_obj
        self.quantity = quantity

    def get_total_price(self):
        return self.product_obj.price * self.quantity

    def __str__(self):
        return f"{self.product_obj.title} | Qty={self.quantity} | Total={self.get_total_price():.2f}"


class Order:
    def __init__(self, id, user_obj, items):
        self.id = id
        self.user_obj = user_obj
        self.items = items


    def get_cart_value(self):
         total = 0
         for item in self.items:
           total += item.get_total_price()
         return total

    def __str__(self):
        item_details = "\n   ".join(str(item) for item in self.items)
        return (
            f"Order ID: {self.id}\n"
            f"User: {self.user_obj}\n"
            f"Items:\n   {item_details}\n"
            f"Cart Value: ₹{self.get_cart_value():.2f}"
        )


#Fetching from the API
def fetch_and_build_data():
    products_data = requests.get("https://fakestoreapi.com/products").json()
    users_data = requests.get("https://fakestoreapi.com/users").json()
    carts_data = requests.get("https://fakestoreapi.com/carts").json()

    #Product mapping
    product_dict = {
        p["id"]: Product(p["id"], p["title"], p["price"], p["category"])
        for p in products_data
    }

    #User mapping
    user_dict = {
        u["id"]: User(u["id"], u["username"], u["email"])
        for u in users_data
    }

    #Orders
    orders = []
    for cart in carts_data:
        user_obj = user_dict.get(cart["userId"])

        items = []
        for item in cart["products"]:
            prod_obj = product_dict.get(item["productId"])
            quantity = item["quantity"]
            items.append(Item(prod_obj, quantity))

        orders.append(Order(cart["id"], user_obj, items))

    return orders


#iterate & print
orders = fetch_and_build_data()

for order in orders:
    print(order)
    print("\n")
    # print("-" * 50)
```
